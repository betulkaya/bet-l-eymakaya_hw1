```python
import thinkstats2
import thinkplot
```

```python
valm=pres.fmarno
hist = thinkstats2.Hist(valm, label='marry')
thinkplot.Hist(hist)
thinkplot.Config(xlabel='yas', ylabel='Pre')
```

```python
nomarried = pres[pres.fmarno == 0]
marrieds = pres[pres.fmarno >0]

nomarriedhist = thinkstats2.Hist(nomarried.totincr, label='nomarried')
marriedshist = thinkstats2.Hist(marrieds.totincr, label='marrieds')
```

```python
width = 0.45
thinkplot.PrePlot(2)
thinkplot.Hist(nomarriedhist, align='right', width=width)
thinkplot.Hist(marriedshist, align='left', width=width)
thinkplot.Config(xlabel='hafta', ylabel='sayi')
```
